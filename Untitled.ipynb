```python
# Create Brownian Motion
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(20,5)})

# Model parameters
t_start = 0
t_end = 200
length = 5000
sigma = 1
r = 2.5

# Define time axis
t = np.linspace(t_start, t_end, length)
dt = np.mean(np.diff(t))

# Initial conditions
x = np.zeros(length)
x[0] = np.random.normal(loc=0.0, scale=1.0)

# Define terms in pitchfork bifurcation
alpha1 = lambda x,t: r*x
alpha2 = lambda x,t: np.power(x,3)
diffusion = lambda x,t: sigma
noise = np.random.normal(loc=0.0,scale=1.0,size=length)*np.sqrt(dt)

# Solve the SDE
for i in range(1,length):
    x[i] = x[i-1] + alpha1(x[i-1], i*dt)*dt - alpha2(x[i-1], i*dt)*dt + diffusion(x[i-1],i*dt)*noise[i]


plt.figure()
ax1 = plt.plot(t,x)
ax2.set(xlabel='days', ylabel='X')

plt.figure()
ax2 = sns.distplot(x,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':0.1})

ax2.set(xlabel='X', ylabel='Frequency')
```
